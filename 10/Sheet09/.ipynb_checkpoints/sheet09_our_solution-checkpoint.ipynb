## Exercise 1: Local features and interest points [2 points]

**a)** Explain in your own words: What are *local features*? How are they used?

YOUR ANSWER HERE  
Local features are features that are smaller and local to an image patch compared to features of a whole image. The aim is to use them to solve common object recognition problems such as partial occlusion, change of or inhomogeneous background, change of illumination, or object transformations (rotation, scaling, transfromation, projectiove /affine transformations).

**b)** What are *interest points* and what are they used for? What properties are desirable?

YOUR ANSWER HERE  
Interest points (IPs) are points that are salient (or they should be as one of the requirements) - which is why they are also called saliency points -, meaning that they are special and uncommon in the respective image or with respect to common images. They are used to find regions (around them) in an image that can then be desribed by suitable features, which in turn can then be used for object recognition.  
Interest points should also be stable, meaning that they keep their position (with respect to the physical world, not in the image) and do not vanish under different conditions/disruptions like illumination changes or a different viewpoint of the same scene

## Exercise 2: Computing interest points [6 points]


**a)** Explain in your own words the idea of the Moravec IP operator. What are its properties? Implement this IP operator and apply it to the image `lighthouse.png`. Try different window sizes and threshold values.

YOUR ANSWER HERE  
The idea behind thee Moravec IP operator is that a window is considered as salient if it is unique in it's surroundings. This uniqueness is found by comparing it to the window shifted by one pixel, which is done in all 4 directions (horizontal, vertical, and both diagonals), and taking the minimum of these 4 values. This is done for all windows in the image, creating a saliency map, which is thresholded to detect corners. So, this is a very easy approach, however it is anisotropic, which means it is not robust against rotation, and it the hard boundaries of the window create problems similar to the box filter.

```python
%matplotlib inline
import imageio
import numpy as np
import scipy.ndimage as nd
import matplotlib.pyplot as plt

img = imageio.imread('images/lighthouse.png', pilmode='F')

window_size = 3
threshold = 0.2

def moravec(img, window_size):
    """Moravec corner detector.
    
    Arguments
    ---------
    img: np.ndarray
        The input image
    window_size: int
        The size of the window to consider

    Results
    -------
    response: np.ndarray
        Response (of the same size as img), indicating interest points.
    """
    #normalize img
    img = img/255.0
    
    response = np.zeros_like(img)
    # YOUR CODE HERE
    # create difference images for all 4 directions
    shift_x = np.zeros_like(img)
    shift_y = np.zeros_like(img)
    shift_diag_up = np.zeros_like(img)
    shif_diag_down = np.zeros_like(img)
    
    shift_x[:-1,:] = img[:-1,:]-img[1:,:]
    shift_y[:,:-1] = img[:,:-1]-img[:,1:]
    shift_diag_up[:-1,1:] = img[:-1,1:]-img[1:,:-1]
    shif_diag_down[:-1,:-1] = img[:-1,:-1]-img[1:,1:]
    
    # square the differences
    shift_x **= 2
    shift_y **= 2
    shift_diag_up **= 2
    shif_diag_down **= 2
    
    # get the shape of the image and calculate how often the window fits in each axis
    y,x = img.shape
    x_windows=x-2*window_size//2+1
    y_windows=y-2*window_size//2+1
    
    # create indexer
    # start with indices for one row of windows in the image
    indexer = np.tile(np.arange(window_size)[None, :], (1, window_size)) + np.arange(x_windows)[:, None]
    shift = np.arange(window_size)*x
    shift = np.repeat(shift, window_size)    
    indexer=indexer+shift
    # then copy and shift that for the each row
    indexer = np.tile(indexer, (y_windows,1))
    shift = np.arange(y_windows)*x
    shift = np.repeat(shift, x_windows)
    shift= shift.reshape((len(shift),1))
    indexer += shift
    
    # use the indexer for all 4 flattened difference arrays
    # each row in the result is one window
    s_x = shift_x.flatten()[indexer]
    s_y = shift_y.flatten()[indexer]
    s_diag_up = shift_diag_up.flatten()[indexer]
    s_diag_down = shif_diag_down.flatten()[indexer]
    
    # sum up the windows
    s_x = np.sum(s_x, axis=1).reshape((s_x.shape[0],1))
    s_y = np.sum(s_y, axis=1).reshape((s_y.shape[0],1))
    s_diag_up = np.sum(s_diag_up, axis=1).reshape((s_diag_up.shape[0],1))
    s_diag_down = np.sum(s_diag_down, axis=1).reshape((s_diag_down.shape[0],1))
    
    # concatonate 4 directions
    windows = np.concatenate((s_x,s_y,s_diag_up,s_diag_down), axis=1)
    # take min of these
    windows = np.min(windows, axis=1)
    # reshape to image (minus borders)
    s_map=windows.reshape((y_windows,x_windows))
    
    
    
    return s_map

response = moravec(img, window_size)

thresholded = np.zeros_like(response)
thresholded[response > threshold] = 1

plt.figure(figsize=(12,8))
plt.subplot(2,2,1); plt.imshow(img, cmap='gray'); plt.title('Original')
plt.subplot(2,2,2); plt.imshow(response, cmap='viridis'); plt.title('Moravec "heatmap"')
plt.colorbar()
plt.subplot(2,2,3); plt.imshow(thresholded, cmap='gray'); plt.title(f'Thresholded (>{threshold})')
plt.subplot(2,2,4); plt.imshow(img[1:-1,:-1], cmap='gray'); plt.title('Corners')
mask = np.zeros(thresholded.shape + (4,), dtype=int)
mask[:,:,0] = 255
mask[:,:,3] = thresholded*255
plt.imshow(mask, interpolation='none')
plt.tight_layout()
plt.show()
```

**b)** How does the Harris corner detector work and in what sense does it improve the Moravec IP operator. Implement the Harris corner detector and apply it to `lighthouse.png`. Again, try different "window sizes" (values for $\sigma$).

YOUR ANSWER HERE    
The Harris corner detector works on the x and y gradients of the image just like the Moravec IP operator, which are extracted in the first step by using the Sobel operator. But instead of shifting a window with a hard boundary, the Harris corner detector uses a so-called structure tensor, that combines the gradients in the neighborhood if a point and consists of a 2x2 matrix with the squared x and y gradients in the diagonal and the mixed producte of the gradients in the corners weighted by the Gaussian as an isotropic windowing function. By using the Gaussian it avoids the porblems of the Moravec operator, as it is isotropic and does not have a hard boundary as the sum is taken over the whole image and the windowing function is used for weighting. 
The last steps in the procedure of the Harris corner detector then are to compute a map containing the edge and corner strenght using the structure tensor and then taking the local maxima of that considering an 8-neighborhood to obtain the corners.

```python
%matplotlib inline
import imageio
import numpy as np
import matplotlib.pyplot as plt
import scipy.ndimage as nd
from skimage.filters import sobel_h, sobel_v

k = .04
window_size = 3

img = imageio.imread('images/lighthouse.png', pilmode='F')/255.

def harris(img, window_size=3, k=0.04):
    """Harris corner detector.
    
    Arguments
    ---------
    img: np.ndarray
        The input image
    window_size: int
        The size of the window to consider
    k: float
        The parameter k
        
    Results
    -------
    response: np.ndarray
        Response (of the same size as img), indicating interest points.
    """
    response = np.zeros_like(img)
    # YOUR CODE HERE
    
    # sobel filter for gradients
    g_x = sobel_h(img)
    g_y = sobel_v(img)
    
    # compute structure tensor    
    A = nd.gaussian_filter(g_x**2, sigma=1)
    C = nd.gaussian_filter(g_x*g_y, sigma=1)
    B = nd.gaussian_filter(g_y**2, sigma=1)
    response = A*B- C**2 - k*(A+B)**2
    
    
    
    # YOUR CODE HERE
    return response
threshold = 0.0007
response = harris(img, window_size, k)

thresholded = np.zeros_like(response)
thresholded[response > threshold] = 1

plt.figure(figsize=(12,8))
plt.subplot(2,2,1); plt.imshow(img, cmap='gray'); plt.title('Original')
plt.subplot(2,2,2); plt.imshow(response, cmap='viridis'); plt.title('Harris "heatmap"')
plt.colorbar()
plt.subplot(2,2,3); plt.imshow(thresholded, cmap='gray'); plt.title(f'Thresholded (>{threshold})')
plt.subplot(2,2,4); plt.imshow(img[1:-1,:-1], cmap='gray'); plt.title('Corners')
mask = np.zeros(thresholded.shape + (4,), dtype=int)
mask[:,:,0] = 255
mask[:,:,3] = thresholded*255
plt.imshow(mask, interpolation='none')
plt.tight_layout()
plt.show()
```

**a)** Explain the two main steps that are typically performed to construct local features. How are these steps realized in SIFT?

YOUR ANSWER HERE  
The two main steps to construct local features are region/IP detection, where ideally salient and stable points and from them regions are detected, and generation of a region descriptor that describes the detected regions using suitable features, which then allow object dtection even through the common problems such as change of illumination or background, object transforms, or partial occlusion.  
In SIFT the detection of keypoints is realized using a scale-space to detect scale adaptive IPs, and the descriptor uses four features - x, y, scale, and direction - the latter of which is obtained using histograms of gradients.

**b)** SIFT key points (interest points) are claimed to invariant to changes in (a) scale, (b) orientation, (c) illumination, and (d) affine transformations. Describe how SIFT key points are computed and explain how this construction contributes to each of these individual goals.

YOUR ANSWER HERE  
Keypoints are computed by creating a scale space where the image is convolved with Gaussians at different scales and subsampled after 2 two convolutions (to save computing resources) and then taking the extrema (minima/maxima) of the difference of the successive gaussian blurred images. The keypoints, then, are those points that occur at multiple scales. As this leads to far too many detected keypoints they are then filtered, removing those keypoints that are unstable for example because they are in low contrast areas or located on edges.  
The descriptor is then constructed using the position and scale of the IP, as well as the direction, which is determined using HOG in area around the IP and taking the most prominent directions, this leads to rotation invarience.
By detecting the keypoints in scale space and automatically extracting the correct scale, which is then saved in the descriptor, the keypoints are invariant to changes in scale. Illumination invariance is achieved by chosing salient points to begin with and sorting out those in low contrast areas that are likely to change if the illuminaiton changes.

## Exercise 5: Rectangle features [3 Points]

**a)** Explain the idea of rectangle features.

YOUR ANSWER HERE  
The idea behind rectangle features is to use a sliding window approach to perform real time object recognition. This is in contrast to interest point approaches, as the window slides over the whole image step by step, not simulating saccades as the biologically inspired IP approach, but instead taking more inspiration from receptive fields in the cortex to which the rectangle features are similar. This approach performs detection by recognition, meaning that at every possible position of the window is classified and thereby - for a positive classification - directly detected, so there is no seperation of localization and classification.  
Getting back to rectangle features, from which this approach gets its name, they are simple features that consist only of black and white rectangular parts and can be interpreted as representing simple geometric features like lines, points, or edges. To compute the feature value, the difference between the integral over the white area and that over the black one is taken. This is done for ervery window and all features and also for different scales by using different window/feature sizes. It is possible to do that by using the integral image which allows performing these computations in very few steps.

**b)** What is an integral image and how does it help to compute rectangle features?

YOUR ANSWER HERE  
An integral image is used in the process of computing rectangle features to speed up the computations involved in that. This is done by precomputing the integral image and then using values from that to compute the feature values, which allows really fast computation and thereby scaling invariance as the number of computations is independent from the window size, so different sizes can be tested as computations are fast enough to allow that.  
To understand how this is possible, one must understand what exactly an intergal image contains. In an integral image each pixel contains the sum of all pixels inside the rectangle spanned up by the origin (being in the upper left corner of the image) and itself. The original image can be efficiently commputed recursively, going through the image just once row by row, as one pixel can be computed from it's own value in the original image and the values of the pixels above, left, and above and left of it in the integral image. 
Once the integral image has been computed one can also quickly calculate the integral for an arbitrary rectancle using just the 4 edge points that define it. This means that instead of having to sum up all the pixels in a window just 4 values have to be lookd up and 3 operations carried out to obtain the result, which is a massive speed up and what makes rectangle features a viable option.

**c)** What is a classifier cascade and how is it used in the context of rectangle features?

YOUR ANSWER HERE  
A classifier cascade, like the name suggests, is a cascade of classifiers, where multiple weak but fast classifiers are are put after each other to build one strong classifier. A weak classifier should have an error rate lower than 50 percent - although higher is also acceptable, as that can be "fixed" by changing the sign and thereby correcting the reversed polarity - and a high detection rate, for which an also high false positive rate is accepted. The high false positive rates will then multiply out over the length of the cascade.  
In the context of rectangle features a classifier cascade is used to build a classifier for the object one wants to detect out of classifiers for individual rectangle features. 

**d)** Compare SIFT and rectangle features. Name pros and cons.

YOUR ANSWER HERE  
SIFT uses an IP based apporach, where recognition and detection happen in two seperate phases, wheares rectangle features use sliding windows that use detection by recognition, so both happen at the same time. The biological motivation for the two also differs accordingly, the detection of few interest points in the image is motivated by saccades while rectangle features can be said to resemble receptive fields in the cortex.   
Both SIFT and rectangle features use simple local features and both are scale invariant, although SIFT is so by design by using scale space during detection and rectangle features obtain that property by testing with many different window sizes wich is a viable option because using the integral image speeds up computations enough.  
SIFT is also rotarion invariant by design as the descriptors contain information about the direction, rectangle features, on the other hand, can achieve rotation invariance only by using rotated rectangle feautures which are hard to generate for angles other than 45 degrees.  
Both methods allow for fast object identification, rectangle features is more efficient as it was specifically designed that way and the gaussians that SIFT uses are harder to compute. SIFT also allows for very precise positioning of IPs with subpixel precision if interpolation is used.
