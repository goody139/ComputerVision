## Assignment 1: Hough transform [7 Points]

**a)** Explain in your own words the idea of Hough transform in general. What is an accumulator space? In what sense can the Hough transform be seen as a model-based approach?

Hough transform detects objects found in lines of geometric figures, like a circle. It detects edge points for example using a sobel filter and transforms those points by pointing in an accumulator space, analysing then the clusters in that specific space. The accumulator space is a space where detected segments are accumulated in a specific representation (depending on the form to be detected); clusters/maxima correspond to the detected form in the image.
HT can be seen as a model-based approach in terms of a simple form whereas the algorithm is looking for small clusters in the accumulator space and therefore specifies the given straight line as the model.Further, this respective straight line is the most simple case which exists if HT was developed. Moreover, it is possible to generalize HT for any forms and is not restricted in any way.
Generally speaking, Hough transform builds on local information, however combines it to become something global which means that overall it may be discussed whether it is local or global in the end.

**b)** What is linear Hough transform? What does a point in the linear Hough space represent? Explain the meaning of the two coordinates.

YOUR ANSWER HERE  
Linear Hough transform is a way to detect straight edges in an image by transforming points in the edge image into lines (representing possible lines through this point) in the accumulator space, estimating the two parameters that define a straight line. The transform space has two dimensions, and every point in the transform space is used as an accumulator to detect or identify a line described. Every point in the detected edges in the image contributes to the accumulators.
Straight lines/edges in the image can then be found by analyzing clusters or "bright spots" in the accumulator space. 

One point in the linear Hough (accumulator) space represents one possible line going through an edge point in the transformed image. The two coordinates in the linear Hough space are r, which is the distance of the line to the origin, and theta, which is the angle of the line that is perpendicular to the represented one and runs through the origin. Each element in the accumulator has a value equal to the sum of the points or pixels that are positioned on the line represented by quantized parameters (r, Œ∏). So the element with the highest value indicates the straight line that is most represented in the input image.

**c)**  How are points, lines, polygons transformed by linear Hough transform? What about parallel lines? Try different configurations using the functions `point`, `line`, `polygon` below. Use the function `skimage.transform.hough_line` to display these examples. You may use the code of Assignment 2 to check
different configurations interactively.

YOUR ANSWER HERE  
A point is transformed into one line in the accumulator space, with the line representing all possible lines (combinations of r and theta) through the point. Consequently a straight line is transformed into multiple lines in the accumulator space, where the intersection of these lines occurs at the r, theta combination of the line in the image. As a line in A is created for every pixel in a line in the image, shorter lines lead to an accumulator space where there are less lines than for longer image lines, and the areas where the lines in the accumulator spaces diverge are also smaller.  
The transformations of parallel lines look similar, but of course the intersections in the accumulator space are  at different rs, and the rest is accordingly different.
Knowing how lines are transformed, polygons are no surprise, as they are made out of lines, the accumulator space just contains the respective representations.

## Assignment 3: Implementing Hough transform [7 Points]

**a)** Implement your own version of linear Hough transform and compare the results with the function `skimage.transform.hough_line`.

```python
from skimage.transform import hough_line
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import warnings
from skimage import filters
from skimage.feature import canny
warnings.filterwarnings('ignore')

def my_hough_line(img, angles=180):
    """
    Apply linear Hough transform to the given image.
    
    Args:
        img (numpy.ndarray): The binary image to transform
        angles (int): The number of angles to consider.
    Result:
        accumulator (numpy.ndarray): The accumulator space.
    """
    # YOUR CODE HERE
    
    # edge point detection using canny
    # TODO if the image is binary we already max have good edges and canny yields only outlines of those
    # so may not be desirable
#     edge_points = canny(img)

    edge_points = img
    
    # we want to compute r for these angles (convert to radians)
    theta = np.linspace(-np.deg2rad(angles/2), np.deg2rad(angles/2), num=angles, endpoint=False)

    # initialize accumulator
    # r size can maximally be as long as the diagonal of the original image    
    r_size = int(np.sqrt(img.shape[0]**2 + img.shape[1]**2)+1)
    
    accumulator = np.zeros(shape=(r_size*2, angles), dtype=np.int64)

    # for all angles to consider
    for angle in theta:
        # only consider edge pixels
        y = np.where(edge_points==True)[0]
        x = np.where(edge_points==True)[1]
        
        # calculate the rs for all edge pixels and the current angle
        r = x*np.cos(angle)+y*np.sin(angle)
        
        # rount r to nearest int and convert to int array to be able to use it as index
        r = np.rint(r)
        r = np.array(r, dtype=np.int64)
        
        # deal with negative numbers to get the proper index
        r[np.where(r<=0)]+=r_size*2
      
        # get the unique idex pairs and the respective counts
        indices = np.vstack((r-r_size, np.repeat(np.round(np.rad2deg(angle))+angles/2, r.size).astype(np.int64)))
        pairs, counts = np.unique(np.transpose(indices), return_counts=True, axis=0)

        # and for all different counts increment the accumulator at the respective indices
        for c in np.unique(counts):
            idxs = np.transpose(pairs[np.where(counts==c)])
            accumulator[idxs[0], idxs[1]] += c
  
    return accumulator


img = np.zeros((100,100))

# You may try different paintings here:
#point(img, (10,10))
#line(img,(10,20),(70,20))
polygon(img,[(20,10),(80,50),(80,80),(30,60)])


fig, [ax1, ax2, ax3] = plt.subplots(1,3, figsize=(12,4))
plt.gray()


# original image
ax1.set_title('Image'); 
ax1.imshow(img, origin = 'lower')

# skimage
out, angles, d = hough_line(img)

out_resized = resize(out, (out.shape[0], out.shape[0]), anti_aliasing=True, preserve_range=True)
ax2.set_title('Hough transform (skimage)');
ax2.set_xlabel('Angles (degrees)')
ax2.set_ylabel('Distance (pixels)')
ax2.imshow(np.log(1 + out_resized), origin = 'lower', cmap='gray')
ax2.set_yticks(np.linspace(0, out.shape[0], 7))
ax2.set_yticklabels((-1 * np.linspace(d[-1], d[0], 7)).astype(int))
ax2.set_xticks(np.linspace(0, out.shape[0], 5))
ax2.set_xticklabels(np.linspace(np.rad2deg(angles[0]), np.rad2deg(angles[-1]), 5).astype(int))

# own implementation
my_out = my_hough_line(img)

my_out_resized = resize(my_out, (out.shape[0], out.shape[0]), anti_aliasing=True, preserve_range=True)
ax3.set_title('Hough transform (own implementation)');
ax3.set_xlabel('Angles (degrees)')
ax3.set_ylabel('Distance (pixels)')
ax3.imshow(np.log(1 + my_out_resized), origin = 'lower', cmap='gray')
ax3.set_yticks(np.linspace(0, out.shape[0], 7))
ax3.set_yticklabels((-1 * np.linspace(d[-1], d[0], 7)).astype(int))
ax3.set_xticks(np.linspace(0, out.shape[0], 5))
ax3.set_xticklabels(np.linspace(np.rad2deg(angles[0]), np.rad2deg(angles[-1]), 5).astype(int))

plt.show()
```

## Assignment 4: Circular Hough [7 Points]

**a)** Explain the idea of circular Hough transform. How many dimensions does the accumulator space have and what do they represent? What practical problems do you expect when computing circular Hough transform?

YOUR ANSWER HERE
The idea of circular Hough transform is to detect circles in (edge) images instead of lines, which is achieved by using a different accumulator space and a accordingly different transformation. Basically for each radius that should be checked a 2d accumulator of the image shape is used, and for each edge point in the image the accumulator space for the currently checked radius in incremented at all pixels that form a circle of the given radius around the center given by the coordinates of the image edge pixel currently checked.  
As can already be read from the previous description, the accumulator space has 3 dimensions, the height, and width (of the image to be transformed), as well as the raddi that one wants to check.  
Practical problems that might occur are a long runtime for large images and multiple radii or that partial/occluded circles might not be detected.

```python
%matplotlib inline
from skimage.transform import hough_line
from scipy import misc
from imageio import imread
import matplotlib.pyplot as plt
import numpy as np

# load the image
img = imread('images/xmas.png')

# YOUR CODE HERE
def circular_hough(img, radii=[10]):
    """
    Apply linear Hough transform to the given image.
    
    Args:
        img (numpy.ndarray): The binary image to transform
        radii (list): The radii to consider.
    Result:
        accumulator (numpy.ndarray): The accumulator space.
    """
    # YOUR CODE HERE
    
    # edge point detection using canny
    edge_points = canny(img)
    
    # initialize accumulator
    accumulator = np.zeros(shape=(img.shape[0], img.shape[1], len(radii)))

    # for all radii to consider
    for r in radii:
        # only consider edge pixels
        xs = np.where(edge_points==True)[0]
        ys = np.where(edge_points==True)[1]
        
        # increment accumulator around edge point in circular fashion
        #for all edge points
        for x, y in zip(xs,ys):
            c = get_circle(img, (y, x) , r)
            accumulator[c] += 1

    return accumulator


def get_circle(img, center, radius):
    result = img.copy()
    h, w = img.shape[:2]

    # np ogrid creates array of indices
    Y, X = np.ogrid[:h, :w]
#     print(X)
#     print((X - center[0]))
#     print()
#     print(Y)
#     print((Y-center[1]))
    
    dist_from_center = np.sqrt((X - center[0])**2 + (Y-center[1])**2)

    e = 0.5
    inner = dist_from_center < radius + e
    outer = dist_from_center > radius - e

    return np.logical_and(inner, outer)


# img = np.zeros((50,50))
# img[get_circle(img, center=(20,20), radius=10)] = 1
# plt.imshow(img)
# plt.show()

# out = circular_hough(img, [11])
# plt.imshow(out)
# plt.show()


plt.figure(figsize=(12, 4))
plt.gray()
plt.subplot(1,3,1) ; plt.title('Image'); plt.imshow(img)

r = 9
out = circular_hough(img, [r])


# out_resized = resize(out, (out.shape[0], out.shape[0]), anti_aliasing=True, preserve_range=True)
ax = plt.subplot(1,3,2)
ax.imshow(out, cmap='gray')

# make a rgb image
path_img = np.zeros((img.shape[0], img.shape[1], 3))
for i in range(3):
    path_img[:,:,i]= img

path_img = path_img.astype(int)
# make path red

ys = np.where(out > np.max(out)*0.7)[0]
xs = np.where(out > np.max(out)*0.7)[1]
for x,y in zip(xs, ys):
    c = get_circle(path_img, center=(x,y), radius=r-1)
    path_img[c, 0] = 255
    path_img[c, 1:] = 0
    
plt.subplot(1,3,3)
plt.imshow(path_img)
plt.show()
```

```python
# imageio
%matplotlib notebook
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import imageio
import numpy as np
from skimage.filters import gaussian
from skimage.feature import canny
from skimage.color import rgb2gray
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.draw import circle_perimeter
from imageio import imread

# we use visvis for a bit better performance compared to matplotlib
import visvis as vv


# List of radii we compute hough transform for. Ideally we adjust for the perfect radius.
hough_radii = [15]
    
def detect_eyes(img):
    """Find pairs of eyes in image based on circular Hough transform.
    
    Args:
        img (ndarray): The RGB input image to detect the eyes in. shape=(h, w, 3)
        
    Returns:
        ndarray: The image with detected eyes highlighted.
    
    """
    # YOUR CODE HERE
    img = img.copy()
    edges = canny(gaussian(rgb2gray(img)))

    hough_res = hough_circle(edges, hough_radii)

    # Select the most prominent 2 circles
    accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii, total_num_peaks=2)
    
    # "draw" them in the return image
    #image = color.gray2rgb(image)
    for center_y, center_x, radius in zip(cy, cx, radii):
        circy, circx = circle_perimeter(center_y, center_x, radius,
                                        shape=img.shape)
        img[circy, circx] = (220, 255, 0)

    return img       
    


try:
    display(HTML("press <kbd>I</kbd>, <kbd>I</kbd> (Kernel Interrupt) to stop the demo!"))
    with imageio.get_reader('<video0>') as webcam:
        t = vv.imshow(webcam.get_next_data(), clim=(0, 255))

        while True:
            img = webcam.get_next_data()
            vv.processEvents()
            # subsample the image also for performance
            img[::2,::2,:] = detect_eyes(img[::2,::2,:])
            t.SetData(img)
except KeyboardInterrupt:
    print("Interrupted")
finally:
    webcam.close()
    plt.close(fig)
    print("Camera was closed.")
    
```
